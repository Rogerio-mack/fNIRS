```python
print(len(df))
df = df.dropna()
print(len(df))
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df.at[23,'ContBlc5_Oc'] = 40
```

```python
for c in df:
  df[c] = df[c].astype('float')
```

```python
desvios = {}
for c in df.drop(columns='Participante'):
  q1 = df[c].quantile(0.25)
  q3 = df[c].quantile(0.75)
  iqr = q3 - q1
  desvios[c] = [ df[ (df[c] < q1 - 1.5*iqr) | (df[c] > q3 + 1.5*iqr) ][c].count() ]

desvios = pd.DataFrame(desvios)
desvios
```

```python
fig, ax = plt.subplots(figsize=(20,5))

i = 0
for c in desvios:
  plt.bar(i,desvios[c],color='b',alpha=0.6)
  i = i + 1

plt.xticks(np.arange(0, len(desvios.columns)), labels=desvios.columns,rotation=90)  

plt.show()
```

```python
desvios_u = {}
for c in df.drop(columns='Participante'):
  q1 = df[c].quantile(0.25)
  q3 = df[c].quantile(0.75)
  iqr = q3 - q1
  for _ , row in df.iterrows():
    # print(int(row['Participante']))
    if (row[c] < q1 - 1.5*iqr) or (row[c] > q3 + 1.5*iqr):
      if int(row['Participante']) in desvios_u.keys():
        desvios_u[int(row['Participante'])] = [ desvios_u[int(row['Participante'])][0] + 1 ]
      else:
        desvios_u[int(row['Participante'])] = [ 1 ] 

desvios_u = pd.DataFrame(desvios_u)
desvios_u
```

```python
desvios_u.columns.sort_values()
```

```python
fig, ax = plt.subplots(figsize=(20,5))

i = 0
for c in desvios_u.columns.sort_values():
  plt.bar(i,desvios_u[c],color='b',alpha=0.6)
  i = i + 1

plt.xticks(np.arange(0, len(desvios_u.columns.sort_values())), labels=desvios_u.columns.sort_values(),rotation=90)  

plt.show()
```
